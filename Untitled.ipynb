```python
import random
import networkx as nx
from collections import OrderedDict
import matplotlib.pyplot as plt
import random
#global str
DG=nx.DiGraph()
edges=[]
s = "w(y)"
def transaction1():
    with open("input1.txt") as fobj:
        text = fobj.read()
        while True:
            try:
                #s = "w(y)"
                if s == "":
                    continue
                if s in text:
                    print(s)
                    DG.add_node("N1")
                    edges.append("N1")
                    break
                raise Exception("No such string found, try again")
            except Exception as e:
                print(e)
def transaction2():
    with open("input2.txt") as fobj:
        text = fobj.read()
        while True:
            try:
                #s = "w(y)"
                if s == "":
                    continue
                if s in text:
                    print(s)
                    DG.add_node("N2")
                    edges.append("N2")
                    break
                raise Exception("No such string found, try again")
            except Exception as e:
                print(e)
def transaction3():
    with open("input3.txt") as fobj:
        text = fobj.read()
        while True:
            try:
                #s = "w(y)"
                if s == "":
                    continue
                if s in text:
                    print(s)
                    DG.add_node("N3")
                    edges.append("N3")
                    break
                raise Exception("No such string found, try again")
            except Exception as e:
                print(e)
def transaction4():
    with open("input4.txt") as fobj:
        text = fobj.read()
        while True:
            try:
                #s = "w(y)"
                if s == "":
                    continue
                if s in text:
                    print(s)
                    DG.add_node("N4")
                    edges.append("N4")
                    break
                raise Exception("No such string found, try again")
            except Exception as e:
                print(e)
    # Get the function from switcher dictionary

#key=int(input("Enter the key value"))
#value=input("Enter the value")

#switcher[key]=value
switcher={
    0:transaction1,
    1:transaction2,
    2:transaction3,
    3:transaction4,
}

while True:
    
    input1=input("would you like to continue y/n")
    if input1=="n":
        break;
    else:
        edges=[]
        number=int(input("enter the number of transaction"))
        keyname=[0,1,2,3]
        for i in range(number):
            n= random.choice([1, 2, 3, 4])

            switcher[keyname[n]]()
    
#print("print the list",edges)

        print(edges)

        DG.add_edges_from([(edges[0],edges[1]),(edges[0],edges[2]),(edges[0],edges[3]),(edges[1],edges[2]),(edges[1],edges[3]),(edges[2],edges[3])])
    
        
        #DG.add_edges_from([("N1","N2"),("N1","N3"),("N1","N4"),("N2","N3"),("N2","N4"),("N3","N4")])
        '''
        if n==4: 
            DG.add_edges_from([(edges[0],edges[1]),(edges[0],edges[2]),(edges[0],edges[3]),(edges[1],edges[2]),(edges[1],edges[3]),(edges[2],edges[3])])
     
        
        
        if n==2:
             DG.add_edges_from([edges[0]],edges[1])
        if n==3:
             DG.add_edges_from([(edges[0])])
        '''
        #DG.add_edge(input2,input3)
        #G.add_edges_from([("a","c"),("c","d"), ("a",1), (1,"d"), ("a",2)])
         
        in_degree=list(DG.in_degree())
        #print(in_degree) 

        print("\n")
        #degree_sequence = sorted([d for n, d in DG.in_degree()], reverse=False)
        degree_sequence = OrderedDict(sorted(in_degree, key=lambda x: x[1]))
        for k, v in degree_sequence.items():
                print(k,v)

        print("\n\n")
        for k, v in degree_sequence.items():
   
            m=str(k)
            if m=="N1":
                print("N1")
                file=open('input1.txt','r')
                for x , N1 in enumerate(file):
                    print(N1)
            if m=='N2':
                print("N2")
                file=open('input2.txt','r')
                for x , N2 in enumerate(file):
                    print(N2)
        
            if m=='N3':
                print("N3")
                file=open('input3.txt','r')
                for x , N3 in enumerate(file):
                    print(N3)
            if m=='N4':
                print("N4")
                file=open('input4.txt','r')
                for x , N4 in enumerate(file):
                    print(N4)
            else:
                print(" ")
        pos = nx.spring_layout(DG)
        nx.draw(DG,pos,node_color='r',node_size=500,alpha=1)
        plt.show()
```
